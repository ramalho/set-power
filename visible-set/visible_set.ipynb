# VisiSet

A set-like collection backed by a numpy structured array with two columns:
`hash` (`int64`) and `value` (`object`).

Jupyter renders instances as numpy-style tables showing the internal hash/value pairs.

```python
from visiset import VisiSet
```

## Examples

```python
VisiSet([0, 1, -1, 2, -2], hash_base=10)
```

```python
VisiSet([0, 1, 7, -2, -1])
```

```python
VisiSet(['açaí', 'banana', 'cherry', 'bananA'], hash_base=2)
```

```python
VisiSet([3.14, True, (1, 2), frozenset({10, 20}), 'Lua'])
```

### Set interface

```python
vs = VisiSet(['a', 'b', 'c'])
vs.add('d')
vs.add('b')  # duplicate — ignored
vs
```

```python
'b' in vs, 'z' in vs
```

```python
len(vs)
```

```python
list(vs)
```

### Plain text repr

```python
print(repr(vs))
```
