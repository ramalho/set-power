# VisiSet

A set-like collection backed by a numpy structured array with two columns:
`hash` (`uint64`) and `value` (`object`).
Jupyter renders instances as numpy-style tables showing the internal hash/value pairs.

```python
from visiset import VisiSet
```

## Examples

```python
vs = VisiSet([0, 1, 2, 3, 4])
vs
```

```python
VisiSet(['apple', 'banana', 'cherry'])
```

```python
VisiSet([3.14, True, (1, 2), frozenset({10, 20})])
```

### Set interface

```python
vs = VisiSet(['a', 'b', 'c'])
vs.add('d')
vs.add('b')  # duplicate — ignored
vs
```

```python
'b' in vs, 'z' in vs
```

```python
len(vs)
```

```python
list(vs)
```

### Plain text repr

```python
print(repr(vs))
```
